# Part II: Scraping the Katz School’s AI Faculty Web Page

```python
url = 'https://www.yu.edu/katz/programs/graduate/artificial-intelligence/faculty'
```

```python
def getDetails(url):
    r = requests.get(url=url)
    # Create a BeautifulSoup object and parse it
    soup = BeautifulSoup(r.text, 'html.parser')

    # get summary text using class = body as per assignment request
    faculty = soup.find("div",{'class':'body'})
    
    # using regular expression get strings of text between the <h5> and <p> separators in HTML
    Nametitle = [head.get_text() for head in faculty.find_all('h5')]
    Background = [head.get_text() for head in faculty.find_all('p')]
    return Nametitle, Background
```

```python
#split the Name and Title string by comma
Nametitle, Background = getDetails(url)
df_Nametitle = pd.DataFrame(Nametitle )[0].str.split(",",expand=True,)

#Create an empty list so we can populate with our list of strings for each of our faculy's
#Biography, Teaching Background
backgroundlist = []
for i in range(int(len(Background)/2)):
    backgroundlist.append(Background[i:i+2])
    i+=2
df_Background = pd.DataFrame(backgroundlist)

#Now we create our final DataFrame with both dataframes of strings above
faculty_info = pd.concat([df_Nametitle, df_Background], axis=1)
faculty_info.columns = ([['Name', 'Title', 'Major','Bio', 'Teaching Background']])
faculty_info
```
